# Text Generation

```python
import os
import requests

os.environ['HF_TOKEN'] = 

API_URL = "https://router.huggingface.co/v1/chat/completions"
headers = {
    "Authorization": f"Bearer {os.environ['HF_TOKEN']}",
}

def query(payload):
    response = requests.post(API_URL, headers=headers, json=payload)
    return response.json()

response = query({
    "messages": [
        {
            "role": "user",
            "content": f'''
                        write the comculsion for input = {unseen_data} and
                        why this output is predicted = {output} ?
                        Explain one by  one input and why  that output?
                        use bullet points for  input  and output
                        Explanation must be single line.
            '''
        }
    ],
    "model": "deepseek-ai/DeepSeek-R1:novita"
})

print(response["choices"][0]["message"])
```
