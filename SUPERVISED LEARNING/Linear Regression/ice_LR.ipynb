```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('C:\\Users\\Admin\\OneDrive\\Desktop\\data science\\ML\\SUPERVISED LEARNING\\Dataset\\Ice Cream Sales - temperatures.csv')
```

```python
X = df[['Temperature']]
y = df['Ice Cream Profits']
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(X,y,train_size=0.8,random_state=42)
```

```python
df.corr()
```

# conclusion
The correlation value is high (around 0.98), so there is a strong relationship.
As temperature increases, ice cream sales also increase.
So we can train a model to predict sales.

```python
model = LinearRegression()

model.fit(xtrain,ytrain) # mx + c
```

```python
model.intercept_
```

```python
model.coef_
```

```python
ytrain_predicted = model.predict(xtrain)
```

```python
ytrain_predicted
```

```python
ytrain - ytrain_predicted 
```

```python
new_input = pd.DataFrame({'Temperature':[1,2,3]})
model.predict(new_input)
```

```python
import matplotlib.pyplot as plt
#train data
#Actual input and output
plt.scatter(xtrain,ytrain)
#Actual input and predicted output
plt.scatter(xtrain,ytrain_predicted)
```

### Train Data

Actual and predicted values are very close. So the model is learning well on the training data.

```python
ytest_predicted=model.predict(xtest)
#test data

#Actual input and output
plt.scatter(xtest,ytest)

#Actual input and predicted output
plt.scatter(xtest,ytest_predicted)
```

### Test Data

Actual and predicted values are very close. So the model is working well on new data.

The model performance is good. Temperature and ice cream sales have strong relationship.
