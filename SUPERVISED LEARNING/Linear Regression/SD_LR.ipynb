```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:\\Users\\Admin\\OneDrive\\Desktop\\data science\\ML\\SUPERVISED LEARNING\\Dataset\\Salary_dataset.csv')
```

```python
X = df[['YearsExperience']]
y = df['Salary']
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(X,y,train_size=0.8,random_state=42)
```

```python
df.corr()
```

```python
model = LinearRegression()

model.fit(xtrain,ytrain) # it finds the pattern from the data # mx+c
```

```python
model.intercept_
```

```python
model.coef_
```

## Predict()  
it accept x value -> it is input value 

it returns predicted value

y = mx + c    -->  =9423.81532303 * x + 24380.20147947369

```python
ytrain_preticted = model.predict(xtrain)
```

```python
ytrain - ytrain_preticted # it returns a actual values and pridicted values differences
```

## Matrics

**used to measure how well model is performing. They tell you whether our model’s predictions are good or bad.**
## R² Score --> How well model captures the pattern (0 to 1)

## MAE --> (Mean Absolute Error)

## MSE --> (Mean Squared Error)

## RMSE --> (Root Mean Squared Error)


```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error,root_mean_squared_error
```

## r2_score

```python
'''
Syntax for all Matrics

Example -

r2_score(actual_output,predicted_output)
r2_score(ytrain, ytrain_predicted)
r2_score(ytest,ytest_predicted)

'''
```

```python
model.score(xtest,ytest)
```

```python
train_score = r2_score(ytrain, ytrain_preticted)
test_score = r2_score(ytest, ytest_predicted)

print(f" Train Score = {train_score}")
print(f" Test Score  = {test_score} ") 
```
