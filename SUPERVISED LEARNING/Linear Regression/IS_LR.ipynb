```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('C:\\Users\\Admin\\OneDrive\\Desktop\\data science\\ML\\SUPERVISED LEARNING\\Dataset\\Ice_cream selling data.csv')
```

```python
X = df[['Temperature (°C)']]
y = df['Ice Cream Sales (units)']
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(X,y,train_size=0.8,random_state=42)
```

```python
df.corr()
```

# Conclusion
The correlation between temperature and ice cream sales is -0.175.

This indicates a very weak negative relationship between the two variables.

so linear regression not working here it predicts but wrong answers only

```python
model = LinearRegression()

model.fit(xtrain,ytrain) #mx+c
```

```python
model.intercept_
```

```python
model.coef_
```

```python
ytrain_predicted=model.predict(xtrain)
```

```python
import matplotlib.pyplot as plt
#train data
#Actual input and output
plt.scatter(xtrain,ytrain)
#Actual input and predicted output
plt.scatter(xtrain,ytrain_predicted)
```

```python
train_score = model.score(xtrain, ytrain)
print("Train Score:", train_score)
```

```python
ytest_predicted=model.predict(xtest)
#test data

#Actual input and output
plt.scatter(xtest,ytest)

#Actual input and predicted output
plt.scatter(xtest,ytest_predicted)
```

```python
test_score = model.score(xtest, ytest)
print("Test Score:", test_score)
```
