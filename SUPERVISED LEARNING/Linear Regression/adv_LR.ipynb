```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score,root_mean_squared_error
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:\\Users\\Admin\\OneDrive\\Desktop\\data science\\ML\\SUPERVISED LEARNING\\Dataset\\Advertising.csv')
```

```python
df.columns
```

```python
df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
df.corr()
```

good correlation in tv and sales and radio and newspaper are weakly correlated. 

model is well in linear regression

```python
X = df.drop(columns=['Sales'])
y = df['Sales']
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(X,y,train_size=0.8, random_state=42)
```

```python
model = LinearRegression()
```

```python
model.fit(xtrain, ytrain)
```

```python
xtrain_score = model.score(xtrain,ytrain)
xtest_score = model.score(xtest,ytest)
print(f'train score: {xtrain_score} \ntest score : {xtest_score}')
```

```python
ytrain_pred = model.predict(xtrain)
ytest_pred = model.predict(xtest)
```

```python
plt.scatter(ytrain, ytrain_pred,color='purple')
plt.scatter(ytest, ytest_pred, color='red')
plt.xlabel('Actual Sales')
plt.ylabel('Predicted Sales')
plt.title('Actual vs Predicted Sales ')
plt.show()
```

```python
mean_absolute_error(ytrain, ytrain_pred)
```

```python
root_mean_squared_error(ytrain, ytrain_pred)
```
