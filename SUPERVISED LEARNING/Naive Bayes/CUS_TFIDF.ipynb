```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

from sklearn.naive_bayes import MultinomialNB
```

```python
df = pd.read_csv('C:\\Users\\Admin\\OneDrive\\Desktop\\data science\\ML\\SUPERVISED LEARNING\\Dataset\\customer_support_tickets.csv')
```

```python
df = df.dropna(axis=1)
```

```python
X = df.drop(columns='Ticket Type')
y = df['Ticket Type']
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(X,y,train_size=0.8, random_state=42)
```

```python
main_pipe = Pipeline(
    steps=[
        ('vector',TfidfVectorizer(stop_words='english',max_features=4000)),
        ('model',MultinomialNB())
    ]
)
```

```python
main_pipe.fit(xtrain,ytrain)
```

```python
ytrain_pred = main_pipe.predict(xtrain)
```

```python
print(classification_report(ytrain,ytrain_pred))
```

```python
ytest_pred = main_pipe.predict(xtest)
print(classification_report(ytest,ytest_pred))
```

**Conclusion**

The model is underfit bcz, train preformance is 51% and test performance is 18% both train test it performs average 


```python
main_pipe = Pipeline( 
    steps=[
        ('vector',CountVectorizer(stop_words='english',max_features=4000)),
        ('model',MultinomialNB())
    ]
)
main_pipe.fit(xtrain,ytrain)
```

```python
ytrain_pred = main_pipe.predict(xtrain)
print(classification_report(ytrain,ytrain_pred))
```
