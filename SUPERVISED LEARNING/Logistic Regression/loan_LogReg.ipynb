```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import confusion_matrix,recall_score,precision_score,accuracy_score,f1_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
df = pd.read_csv('C:\\Users\\Admin\\OneDrive\\Desktop\\data science\\ML\\SUPERVISED LEARNING\\Dataset\\loan_approval_dataset.csv')
df
```

```python
X = df.drop(columns = 'Loan_Approved')
y = df['Loan_Approved']
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(X,y,random_state=42,train_size=0.8)
```

```python
cat_cols = X.select_dtypes(include = 'object').columns
```

```python
encoder = OneHotEncoder(handle_unknown='ignore',sparse_output=False)
encoded_values = encoder.fit_transform(xtrain[cat_cols])
columns = encoder.get_feature_names_out()
xtrain[columns] = encoded_values
xtrain.drop(columns=cat_cols,inplace=True)
```

```python
xtrain
```

```python
encoded_values_test = encoder.transform(xtest[cat_cols])
columns = encoder.get_feature_names_out()
xtest[columns] = encoded_values_test
xtest.drop(columns=cat_cols,inplace=True)
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(xtrain,ytrain)
```

## **TRAIN DATA**

```python
ytrain_pred = model.predict(xtrain)
print(classification_report(ytrain,ytrain_pred))
```

**Conclusion**

* in that **predicted values** currectly predicted values as No --> 54% and falsely predicted as No -> 46%

currectly predicted as Yes --> 55% and falsely predicted as Yes -> 45% 

* in **actual values** currectly predicted as No --> 51% and falsely predicted as No -> 49%

currectly predicted as Yes --> 58% and falsely predicted as Yes -> 42%

* in the All samples currectly predicted both Yes and No as --> 55% 

## **TEST DATA**

```python
ytest_pred = model.predict(xtest)
print(classification_report(ytest,ytest_pred))
```
