```python
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
import pandas as pd
```

```python
df = pd.read_csv('C:\\Users\\Admin\\OneDrive\\Desktop\\data science\\ML\\SUPERVISED LEARNING\\Dataset\\loan_approval_dataset.csv')
X = df.drop(columns='Loan_Approved')
y = df['Loan_Approved']
df
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(X,y,train_size=0.8,random_state=42)
```

```python
num_cols = X.select_dtypes(include='number').columns
obj_cols = X.select_dtypes(exclude='number').columns
```

```python
preprocessing = ColumnTransformer(
    transformers=[
        ('scaling',StandardScaler(),num_cols),
        ('encoder',OneHotEncoder(),obj_cols)
    ],
    remainder='passthrough'
)
values = preprocessing.fit_transform(xtrain)
```

```python
col_name = preprocessing.get_feature_names_out()
```

```python
xtrain = pd.DataFrame(data=values,columns=col_name)
xtrain
```

```python
xtest = preprocessing.transform(xtest)
```
