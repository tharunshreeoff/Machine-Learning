```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, OrdinalEncoder
from category_encoders import TargetEncoder
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv('C:\\Users\\Admin\\OneDrive\\Desktop\\data science\\ML\\SUPERVISED LEARNING\\Dataset\\employee_attrition_dataset.csv')
df
```

```python
X = df.drop(columns=['Monthly_Income'])
y = df['Monthly_Income']
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(X,y,train_size=0.8,random_state=42)
```

## Label Encoder

```python
encoder = LabelEncoder()
xtrain['Marital_Status'] = encoder.fit_transform(xtrain['Marital_Status'])
xtest['Marital_Status'] = encoder.transform(xtest['Marital_Status'])
```

```python
print(encoder.transform(['Married']))
print(encoder.transform(['Single']))
encoder.transform(['Divorced'])
```

## OneHot Encoder

```python
encoder = OneHotEncoder(sparse_output=False,handle_unknown='ignore')

xtrain[['Gender_Female','Gender_Male']] = encoder.fit_transform(xtrain[['Gender']])
xtrain.drop(columns='Gender',inplace=True)
```

```python
encoder.transform(xtest[['Gender']])
# xtest.drop(columns='Gender',inplace=True)
```

```python
encoder.transform([['Female'],['Male'],['OK']])
```

## Ordinal Encoder

```python
xtrain['Department'].unique()
```

```python
encoder = OrdinalEncoder(categories=[['HR','IT','Sales','Marketing','Finance']],handle_unknown='use_encoded_value',unknown_value=-1)
xtrain['Department'] = encoder.fit_transform(xtrain[['Department']])
xtest['Department'] = encoder.transform(xtest[['Department']])
```

```python
encoder.transform([['IT'],['Sales'],['Finance'],['HR'],["Ok"]])
```
