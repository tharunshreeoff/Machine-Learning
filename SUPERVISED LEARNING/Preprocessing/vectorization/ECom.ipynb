```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv(
    "C:\\Users\\Admin\\OneDrive\\Desktop\\data science\\ML\\SUPERVISED LEARNING\\Dataset\\ecommerceDataset.csv",
    header=None,
    names=["Category", "Description"]
)

df.head()
```

```python
df.dropna(inplace=True)
df.info()
```

```python
X = df.Category
y = df.Description
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(X,y,random_state=42,train_size=0.8)
```

```python
main_pipe = Pipeline(
    steps=[
        ('vector',CountVectorizer()),
        ('model',LogisticRegression())
    ]
)
main_pipe.fit(xtrain,ytrain)
```
