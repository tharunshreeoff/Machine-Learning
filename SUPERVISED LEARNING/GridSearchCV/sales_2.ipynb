```python
import pandas as pd
from sklearn.model_selection import train_test_split
from category_encoders import TargetEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error,root_mean_squared_error,r2_score
```

```python
df = pd.read_csv('C:\\Users\\Admin\\OneDrive\\Desktop\\data science\\ML\\SUPERVISED LEARNING\\Dataset\\sales_data.csv')
df =df.drop(columns=['Region','Sales_Rep','Sale_Date'])
df.duplicated().any()
df['Product_ID'] = df['Product_ID'].astype('str')
```

```python
X = df.drop(columns='Sales_Amount')
y = df.Sales_Amount
```

```python
num_cols = X.select_dtypes(include='number').columns
cat_cols = X.select_dtypes(exclude='number').columns
# X[cat_cols] = X[cat_cols].astype('category')
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(X,y,train_size=0.8,random_state=42)
```

```python
cat_cols
```

```python
# encoder = TargetEncoder(cols='Product_ID')
# xtrain = encoder.fit_transform(xtrain,ytrain)
```

```python
preprocessing = ColumnTransformer(
    transformers=[
        ('encode',OneHotEncoder(),['Product_Category', 'Customer_Type', 'Payment_Method','Sales_Channel']),
        ('encoder',TargetEncoder(),['Product_ID','Region_and_Sales_Rep'])
    ],remainder='passthrough'
)
main_pipe = Pipeline(
    steps=[
        ('pre',preprocessing),
        ('model',RandomForestRegressor())
    ]
)
# main_pipe.fit(xtrain,ytrain)
```

```python
grid_cv = GridSearchCV(
    estimator=main_pipe,
    param_grid={'model__criterion' : ['squared_error','absolute_error'],
                'model__max_depth':[None,5,10,15,50,100],
                'model__min_samples_split':[2,5,7,10,15],
                'model__min_samples_leaf':[1,3,5,7,10],
                'model__max_features':['sqrt',1]},
    cv=5,
    n_jobs=-1,
    verbose=2
)
grid_cv.fit(xtrain,ytrain)
```

```python
grid_cv.best_estimator_
```

```python
res = grid_cv.cv_results_
pd.DataFrame(res)
```

```python
ytrain_pred = grid_cv.predict(xtrain)
```

```python
r2_score(ytrain,ytrain_pred)
```

```python
ytest_pred = grid_cv.predict(xtest)
r2_score(ytest,ytest_pred)
```
